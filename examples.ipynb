# Imports

```python
%matplotlib inline
```

```python
import numpy as np

import matplotlib.pyplot as plt
```

```python
import streamlines
```

# Setup Dummy Data

```python
r = np.linspace ( 1., 0., 256 )
theta = np.linspace( 0., 6.*np.pi, 256)
```

```python
x = r * np.cos( theta )
y = r * np.sin( theta )
```

# Plot

```python
fig = plt.figure( figsize=(8,8), facecolor='white' )
ax = plt.gca()

streamlines.streamline_plot(
    ax,
    x,
    y,
)

ax.set_xlim( [-1, 1 ])
ax.set_ylim( [-1, 1 ])
```

```python
fig = plt.figure( figsize=(8,8), facecolor='white' )
ax = plt.gca()

streamlines.streamline_plot(
    ax,
    x,
    y,
    linewidth = 10,
    color1 = '#22B200',
    color2 = '#FF1DCB',
    color_power = 0.5,
)

ax.set_xlim( [-1, 1 ])
ax.set_ylim( [-1, 1 ])
```

```python
fig = plt.figure( figsize=(8,8), facecolor='white' )
ax = plt.gca()

for scale in np.linspace( 0., 5., 20 ):
    
    streamlines.streamline_plot(
        ax,
        x*scale,
        y*scale,
    )

ax.set_xlim( [-1, 1 ])
ax.set_ylim( [-1, 1 ])
```
